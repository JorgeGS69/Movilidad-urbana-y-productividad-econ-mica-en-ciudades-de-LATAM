### 2.3 Corregir formatos num√©ricos y de fecha

**üéØObjetivo:**
Asegurar que las columnas de fechas y valores num√©ricos est√©n en formatos correctos para permitir an√°lisis, c√°lculos y comparaciones precisas.

**Instrucciones:**

- Convierte las columnas de fecha de `traffic` a formato `datetime`. Haz el cambio a prueba de errores.
- En el dataset `eco`, limpia los valores num√©ricos:
    - En `city_gdp_capita`: elimina separadores de miles (`.`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
    - En `unemployment_pct`: elimina el s√≠mbolo de porcentaje (`%`) y reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
    - En `population_m`: reemplaza las comas (`','`) por puntos (`'.'`) antes de convertir a tipo `float`.
- Finalmente, crea una nueva columna llamada `population` multiplicando `population_m` por 1,000,000 para obtener la poblaci√≥n total.


<details>
<summary>Haz clic para ver la pista</summary>
para eliminar s√≠mbolos, puedes reemplazarlos por un texto vac√≠o.

```python
# Convertir las columnas de traffic a tipo fecha con pd.to_datetime()
traffic['update_time_utc'] = pd.to_datetime(traffic['update_time_utc'], errors="coerce")
 #tu c√≥digo aqu√≠
traffic['update_time_utc_week_ago'] = pd.to_datetime(traffic['update_time_utc_week_ago'], errors="coerce")
#tu c√≥digo aqu√≠

# verificar el cambio
traffic.info()
```

<div class="alert alert-block alert-success">
<b>Comentario de Revisor          </b> <a class="tocSkip"></a>


Correcto, bien con el uso de `pd.to_datetime` para realizar la conversi√≥n a datetime de las columnas que contienen fechas. Bien al agregar `errors='coerce'` como par√°metro, si hay una fila que no est√° en formato fecha, quedar√° como nulo.


</div>

```python
# Limpia separadores y convierte columnas num√©ricas en eco
# eco['city_gdp_capita'] = ....astype(str).str.replace('.', '').str.replace(',', '.').astype(float)
eco['city_gdp_capita'] = eco['city_gdp_capita'] = (
       eco['city_gdp_capita']          # seleccionamos la columna unit_2
     .astype("string")     # convertimos a string
     .str.replace('.', '') #reemplazamos comas por nada
     .str.strip()          # quitamos espacios al inicio y final
     .str.replace(",", ".")   #reemplazamos comas por puntos
     .astype(float)
)
# eco['unemployment_pct'] = ...
eco['unemployment_pct'] = eco['unemployment_pct'] = (
       eco['unemployment_pct']          # seleccionamos la columna unit_2
     .astype("string")     # convertimos a string
     .str.strip()          # quitamos espacios al inicio y final
     .str.replace(",", ".")   #reemplazamos comas por puntos
     .str.replace('%', "")   #reemplazamos % por puntos      
     .astype(float)
)
# eco['population_m'] = ...
eco['population_m'] = eco['population_m'] = (
       eco['population_m']          # seleccionamos la columna unit_2
     .astype("string")     # convertimos a string
     .str.strip()          # quitamos espacios al inicio y final
     .str.replace(",", ".")   #reemplazamos comas por puntos
     .astype(float)
)
# eco['pm2_5'] = ...
eco['pm25'] = eco['pm25'] = (
       eco['pm25']          # seleccionamos la columna unit_2
     .astype("string")     # convertimos a string
     .str.strip()          # quitamos espacios al inicio y final
     .str.replace(",", ".")   #reemplazamos comas por puntos
     .astype(float)
)


# Calcula la poblaci√≥n total en unidades absolutas (Multiplica * 1000000)
# eco['population'] = ...
eco['population'] = eco['population_m'] * 1000000


# verificar el cambio
eco.info()
eco.head(3)
```

## üß©Paso 4: Analizar y resumir datos de movilidad

Como el dataset de tr√°fico contiene **m√∫ltiples registros por ciudad**. En esta parte, calcular√°s los promedios anuales por ciudad para simplificar el an√°lisis y obtener una visi√≥n m√°s clara de las tendencias generales.

### 4.1 Calcular promedios de tr√°fico por ciudad

**üéØObjetivo:**
Obtener una vista consolidada del tr√°fico promedio por ciudad y a√±o, para analizar patrones generales sin depender de datos diarios.

**Instrucciones**

- Agrupa los datos por `city`, `country` y `year`.
- Calcula el promedio **solo de las m√©tricas de tr√°fico m√°s relevantes**: como `jams_delay`, `traffic_index_live`, `jams_length_kms`, `jams_count`, `mins_delay`, y tiempos de viaje (`travel_time_live_per_10kms_mins` y `travel_time_hist_per_10kms_mins`).
- Guarda el resultado como `traffic_city_year_2024`, mant√©n las columnas como variables (no √≠ndices).


<details>
<summary>Haz clic para ver la pista</summary>
Usa ".agg()" para aplicar funciones de promedio. Al final, reinicia el √≠ndice para mantener las columnas de la agrupaci√≥n como variables (no √≠ndices).

```python
# Calcular los  promedios de trafico por ciudad, pa√≠s y a√±o
group_cols = ['city', 'country', 'year']
traffic_city_year_2024 =  traffic_2024.groupby(group_cols, dropna=False).agg({'jams_delay':'mean',
                                                                              'traffic_index_live':'mean',
                                                                              'jams_length_kms':'mean',
                                                                              'jams_count':'mean',
                                                                              'mins_delay':'mean',
                                                                              'travel_time_live_per_10kms_mins':'mean',
                                                                              'travel_time_hist_per_10kms_mins':'mean'})
# tu c√≥digo aqui

# Mostrar resultado
traffic_city_year_2024.head()
```

<div class="alert alert-block alert-warning">
<b>Comentario de Revisor          </b> <a class="tocSkip"></a>

Muy bien con los resultados, la agrupaci√≥n se realiz√≥ por las columnas correctas y se calcul√≥ la media adecuada. 

Sin embargo, hay una mejora sutil que podr√≠as realizar. Al realizar agrupaciones con `groupby()` en pandas, el resultado normalmente deja a las columnas de agrupaci√≥n como √≠ndice. Esto quiere decir que en este caso city, country y year no son columnas normales del dataframe, sino que en realidad son parte del √≠ndice del dataframe. Por ejemplo, si quisieras usar la columna country y haces `traffic_city_year_2024['country']`, obtendr√°s un error. 

Para dejar esas columnas como columnas "normales" del dataframe puedes hacerlo de dos formas:

- Al realizar una agrupaci√≥n, usa el par√°metro `as_index=False` dentro de `groupby`, en este caso podr√≠as haber hecho `.groupby(['city','country','year'], as_index=False)`.
- Al terminar la agrupaci√≥n, si no usaste `as_index=False` y quieres dejar como columnas normales las columnas que quedaron como √≠ndice, puedes hacer `reset_index()`. Esto, modifica el √≠ndice del dataframe y lo deja como un √≠ndice num√©rico correlativo (va de 0 a la cantidad de filas del dataframe) y por defecto mueve las columnas que estaban como √≠ndice con el resto de las columnas.
